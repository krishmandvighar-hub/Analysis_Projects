```python
#🔹 Basic Python Questions
#1. Load the dataset using Pandas.
import pandas as pd 
import matplotlib.pyplot as plt

df=pd.read_csv("products-1000.csv")

print(df)
```

```python
# 2. Find total rows and columns.
print(df.shape)
```

```python
# 3. Display all column names.
print(df.columns)
```

```python
# 4. Check missing values.
print(df.isnull().sum())
```

```python
# 5. Find data types of all columns.
print(df.dtypes)
```

```python
# 🔹 NumPy Questions
# 6. Convert a numeric column into NumPy array.
import numpy as np
price_array=np.array(df["Price"])
print(price_array)
```

```python
# 7. Find mean, median, and standard deviation.
print("MEAN :",np.mean(price_array))
print("MEDIAN :",np.median(price_array))
print("STANDARD :",np.std(price_array))
```

```python
# 8. Find maximum and minimum values.
print(np.max(price_array))
print(np.min(price_array))
```

```python
# 9. Find products with price greater than average price.
avg=np.mean(price_array)
print(df[df["Price"]>avg])
```

```python
# 🔹 Pandas Analytics Questions
# 10. Find top 10 expensive products.
print(df[["Price","Name"]].sort_values(by="Price",ascending=False).head(10))
```

```python
# 11. Count products category-wise.
pro_cat=df["Category"].value_counts()
print("Count products category:",len(pro_cat))
```

```python
# 12. Find average price by category.
pri_cat=df.groupby("Category")["Price"].mean()
print(pri_cat)
```

```python
# 13. Find total stock by category.
sto_cat=df.groupby("Category")["Stock"].sum()
print(sto_cat)
```

```python
# 14. Find duplicate rows.
print(df.duplicated().sum())
```

```python
# 15. Count unique categories
print(df['Category'].nunique())
```

```python
# 16. Find unique categories
print(df["Category"].unique())
```

```python
# 17. Find total products in each brand
print(df['Brand'].value_counts())
```

```python
# 🔹 Matplotlib Visualization Questions
# 15. Create bar chart for category counts.
df["Category"].value_counts().plot(kind="bar")
plt.title("Category Count")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
# 16. Create histogram of prices.
df["Price"].plot(kind="hist")
plt.title("Price Distribution")
plt.xlabel("Price")
plt.show()
```

```python
# 17. Create pie chart of top categories.
df['Category'].value_counts().head(5).plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.ylabel('')
plt.title("Top Categories")
plt.show()
```
