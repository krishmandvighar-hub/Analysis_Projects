```python
import pandas as pd
df=pd.read_csv("blinkit_store_data.csv")
print(df.shape)
print(df.columns)
print(df.info())
```

```python
#1. Find the top 10 cities by total revenue.
top10_city=df.groupby("City")["Revenue"].sum().sort_values(ascending=False).head(10)
print(top10_city)
#optional
top10_city.plot(kind="line",marker="o")
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Which product category generated the highest sales?
product_category=df.groupby("Category")["Revenue"].sum().sort_values(ascending=False)
print(product_category.head(1))
```

```python
# 3. Calculate average delivery time city-wise.
Time_City=df.groupby("City")["Delivery_Time_Minutes"].mean().sort_values(ascending=False)
print(Time_City)
```

```python
# 4. Find the total number of orders for each payment method.
Total_orders=df["Payment_Method"].value_counts().sort_values(ascending=False)
print(Total_orders)
```

```python
# 5. Identify stores with the highest average customer rating.
High_avg_rating=df["Customer_Rating"].mean()
print("highest average customer rating: ",High_avg_rating)
```

```python
# 6. Analyze monthly revenue trends using line charts.
import matplotlib.pyplot as plt
df["date_line"]=pd.to_datetime(df["Order_Date"]).dt.month_name()
# print(date_line)
trends=df.groupby("date_line")["Revenue"].sum().sort_values(ascending=False)
print(trends)
trends.plot(color="orange",marker="o")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
# 7. Find the top 5 products by quantity sold.
Quantity_sold=df.groupby("Product_Name")["Quantity"].count().sort_values(ascending=False).head(5)
print(Quantity_sold)
#optional
Quantity_sold.plot(kind="line",marker="o")
plt.show()
```

```python
# 8. Calculate cancellation and return percentages.
Delivery=df["Delivery_Status"].value_counts().reset_index()
divi=100/Delivery['count'].sum()
print(divi)
delivery_per=Delivery['count'].multiply(divi)
print(delivery_per)
print(Delivery)
```

```python
plt.bar(Delivery['Delivery_Status'],delivery_per,color="r")
plt.xlabel("Delivery Status ")
plt.ylabel("Status_count %")
plt.show()
```

```python
# 9. Compare average revenue per order across cities.
avr_order_bycity=df.groupby("City")["Revenue"].mean().sort_values(ascending=False)
print(avr_order_bycity)
#optional
avr_order_bycity.plot(kind="line",marker="o")
plt.show()
```

```python
# 10. Create a bar chart of category-wise revenue.
product_category.plot(kind="bar",color="c")
plt.title("category-wise revenue")
plt.ylabel("Revenue")
plt.show()
```

```python
# 11. Create a histogram for delivery times.
df.plot(kind="hist",y="Delivery_Time_Minutes")
plt.show()
```

```python
# 12. Analyze relationship between discount percentage and revenue.
correlation=df['Discount_Percent'].corr(df['Revenue'])
print("Correlation: ",correlation)
```

```python
# 13. Find peak ordering months and weekdays.
df["month_line"]=pd.to_datetime(df["Order_Date"]).dt.month_name()
print(df["month_line"].value_counts().head(1))

df["order_line"]=pd.to_datetime(df["Order_Date"]).dt.day_name()
print( df["order_line"].value_counts().head(1))

# df['Order_Date'] = pd.to_datetime(df['Order_Date'])
# monthly_orders = df.groupby(df['Order_Date'].dt.month).size()
# print(monthly_orders.sort_values(ascending=False))

```

```python
# 14. Create pie chart for payment method distribution.

Total_orders.plot(kind="pie",autopct="%1.1f%%")
plt.show()
```

```python
# 15. Identify products with highest discounts.
High_pro=df.groupby("Product_Name")["Discount_Percent"].max().sort_values(ascending=False)
print(High_pro.head(1))
#optional
High_pro.head(5).plot(kind="bar")
plt.show()
```

```python
# 16. Compare delivery performance across stores.
store_per=df.groupby("Store_ID")["Delivery_Time_Minutes"].mean().sort_values(ascending=False)
print(store_per)
#optional
store_per.plot(kind="bar")
plt.show()
```

```python
# 17. Create scatter plot between customer rating and delivery time
plt.scatter(df["Customer_Rating"],df["Delivery_Time_Minutes"])
plt.title("Customer Rating vs Delivery Time")
plt.xlabel("Customer Rating")
plt.ylabel("Delivery Time (Minutes)")
plt.show()
```

```python
# 18. Find the most profitable store in each city.
profitable_store=df.groupby("Store_ID")["Revenue"].sum().sort_values(ascending=False)
print(profitable_store)
#optional
profitable_store.plot(kind="bar")
plt.show()
```

```python
# 19. Analyze revenue contribution by delivery status.
delivery_revenue = df.groupby('Delivery_Status')['Revenue'].sum()

print(delivery_revenue)
delivery_revenue.plot(kind='pie', autopct='%1.1f%%', figsize=(7,7))

plt.title("Revenue Contribution by Delivery Status")
plt.ylabel('')
plt.show()
```
